```python
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, Latex
import numpy as np
import json
import pandas as pd
from scipy import stats

with open('mapping_experiment.json', 'r') as fp:
    data = json.load(fp)
```

```python
experiments = [
            {'id': '1', 'source':'imdb', 'target':'uwcse', 'predicate':'workedunder', 'to_predicate':'advisedby'},
            {'id': '2', 'source':'uwcse', 'target':'imdb', 'predicate':'advisedby', 'to_predicate':'workedunder'},
            #{'id': '3', 'source':'yeast', 'target':'twitter', 'predicate':'interaction', 'to_predicate':'follows'},
            #{'id': '4', 'source':'twitter', 'target':'yeast', 'predicate':'follows', 'to_predicate':'interaction'},
            {'id': '5', 'source':'imdb', 'target':'cora', 'predicate':'workedunder', 'to_predicate':'samevenue'},
            {'id': '6', 'source':'uwcse', 'target':'cora', 'predicate':'advisedby', 'to_predicate':'samevenue'},
            {'id': '7', 'source':'imdb', 'target':'cora', 'predicate':'workedunder', 'to_predicate':'sameauthor'},
            ]

#display(Markdown('# Results for ' + metric))
table = []
for j in range(len(experiments)):
    dataset = experiments[j]['id'] + '_' + experiments[j]['source'] + '_' + experiments[j]['target']
    if dataset in data['results']: 
        found_aucroc = np.array([item['Evaluating all']['Found AUC ROC'] for item in data['results'][str(dataset)]])
        found_aucpr = np.array([item['Evaluating all']['Found AUC PR'] for item in data['results'][str(dataset)]])
        found_cll = np.array([item['Evaluating all']['Found CLL'] for item in data['results'][str(dataset)]])
        best_aucroc = np.array([item['Evaluating all']['Best AUC ROC'] for item in data['results'][str(dataset)]])
        best_aucpr = np.array([item['Evaluating all']['Best AUC PR'] for item in data['results'][str(dataset)]])
        best_cll = np.array([item['Evaluating all']['Best CLL'] for item in data['results'][str(dataset)]])
        time_spent = np.array([item['Evaluating all']['Time spent'] for item in data['results'][str(dataset)]])
        alg_time = np.array([item['Mapping results']['Total time'] for item in data['results'][str(dataset)]])
        pmap = np.array([item['Mapping results']['Possible mappings'] for item in data['results'][str(dataset)]])
        
        aucroc_ratio = found_aucroc.mean() / best_aucroc.mean()
        aucpr_ratio = found_aucpr.mean() / best_aucpr.mean()
        cll_ratio = 0.0 #found_cll.mean() / best_cll.mean()
        time_ratio = alg_time.mean() / time_spent.mean()
        pmap = pmap.mean()

        table.append([dataset, '%.3f' % (aucroc_ratio),
        '%.3f' % (aucpr_ratio),
        '%.3f' % (cll_ratio),
        '%.3f' % (time_ratio),
        '%d' % (pmap),])
display(pd.DataFrame(table, columns=['Experiment', 'AUC ROC ratio', 'AUC PR ratio', 'CLL ratio', 'Time ratio', 'Possible mappings']))
```
