```python
seed = 441773
```

```python
print('Number of possible groundings')

for experiment in experiments:
    target = experiment['target']
    to_predicate = experiment['to_predicate']
    balanced = 1
    tar_data = datasets.load(target, bk[target], target=to_predicate, balanced=balanced, seed=seed)
    print('\nFor experiment ' + experiment['id'] + '_' + experiment['source'] + '_' + experiment['target'] + ', target is ' + experiment['to_predicate'])
    if target in ['nell_sports', 'nell_finances']:
        tar_data[1] = datasets.split_into_folds(tar_data[1][0], n_folds=3, seed=seed)
    for i in range(len(tar_data[1])):
        fold = tar_data[1][i]
        print('Fold ' + str(i+1))
        set1 = set()
        set2 = set()
        for example in fold:
            match = re.match('([a-zA-Z_0-9]*)\s*\(([a-zA-Z_0-9,\s]*)\).', example)
            if match:
                args = match.group(2).split(',')
                set1.add(args[0])
                set2.add(args[1])
        if to_predicate in ['publication', 'genre', 'bankchiefexecutiveceo', 'companyeconomicsector', 'teamplayssport']:
            print(str(len(set1)) + ' x ' + str(len(set2)) + ' = ' + str(len(set1) * len(set2)))
        else:
            print(str(len(set1.union(set2))) + ' x ' + str(len(set1.union(set2))) + ' = ' + str(len(set1.union(set2)) * len(set1.union(set2))))
```
